B - Commencement

```python
s = input()
chars = [0] * 26
base = ord("a")
for c in s:
    chars[ord(c) - base] += 1
for i in range(1,len(s)+1):
    acc = 0
    for c in chars:
        if c == i:
            acc += 1
    if acc != 0 and acc != 2:
        print("No")
        exit()
print("Yes")
```

C - Airport Code

```python
s = input().upper()
t = input()
i = 0
j = 0
if t[2] == "X":
    nt = 2
else:
    nt = 3
while i < len(s):
    if s[i] == t[j]:
        j += 1
    if j == nt:
        print("Yes")
        sys.exit()
    i += 1
print("No")
```
