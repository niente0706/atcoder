C - Grand Garden

```python
n = int(input())
hls = list(map(int, input().split()))

def list_split(ls):
    result = [min(ls)]
    acc = []
    for i in range(len(ls)):
        if ls[i] == result[0]:
            if acc != []:
                result.append(acc)
                acc = []
        else:
            acc.append(ls[i] - result[0])
    if acc != []:
        result.append(acc)
    return result

n = 0
queue = [hls]
while queue != []:
    ls_split = list_split(queue[0])
    n += ls_split[0]
    queue = queue[1:] + ls_split[1:]

print(n)
```
