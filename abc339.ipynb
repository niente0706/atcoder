B - Langton's Takahashi

```python
h,w,n = map(int, input().split())
field = []
for _ in range(h):
    field.append(["."] * w)
move = [[-1,0],[0,1],[1,0],[0,-1]]
takahashi = [0,0,0]
for _ in range(n):
    if field[takahashi[0]][takahashi[1]] == ".":
        takahashi[2] = (takahashi[2] + 1) % 4
        field[takahashi[0]][takahashi[1]] = "#"
    else:
        takahashi[2] = (takahashi[2] - 1) % 4
        field[takahashi[0]][takahashi[1]] = "."
    takahashi[0] = (takahashi[0] + move[takahashi[2]][0]) % h
    takahashi[1] = (takahashi[1] + move[takahashi[2]][1]) % w
for i in range(h):
    print("".join(field[i]))
```

C - Perfect Bus

```python
n = int(input())
alist = list(map(int, input().split()))
acc = 0
imin = -1
amin = 0
for i in range(n):
    acc += alist[i]
    if acc <= amin:
        amin = acc
        imin = i
print(sum(alist[imin+1:]))
```
