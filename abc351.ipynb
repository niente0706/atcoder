B - Spot the Difference

```python
n = int(input())
i = 1
a = []
while i <= n:
    a.append(input())
    i += 1
i = 1
while i <= n:
    bi = input()
    for j in range(n):
        if a[i-1][j] != bi[j]:
            print(i,j+1)
            exit()
    i += 1
```

C - Merge the balls

```python
n = int(input())
alist = list(map(int, input().split()))
n = 0
acc = [alist[0]]
for a in alist[1:]:
    if a > acc[-1]:
        n += len(acc)
        acc = [a]
    else:
        acc.append(a)
    while len(acc) > 1 and acc[-1] == acc[-2] :
        add = acc.pop() + 1
        acc.pop()
        acc.append(add)
print(n + len(acc))
```
