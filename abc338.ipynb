A - Capitalized?

```python
s = input()
if s[0].islower():
    print("No")
    exit()
if len(s) >= 2:
    for i in range(1,len(s)):
        if s[i].isupper():
            print("No")
            exit()
print("Yes")
```

B - Frequency

```python
s = input()
clist = [0] * 26
base = ord("a")
for c in s:
    clist[ord(c) - base] += 1
m = max(clist)
for i in range(26):
    if clist[i] == m:
        print(chr(base + i))
        exit()
```

C - Leftover Recipes

```python
n = int(input())
qlist = list(map(int, input().split()))
alist = list(map(int, input().split()))
blist = list(map(int, input().split()))

allmax = 0
i = 0
flag = True
while flag:
    bmax = 10 ** 7
    for j in range(n):
        if qlist[j] - (alist[j] * i) < 0:
            print(allmax)
            exit()
        if blist[j] == 0:
            continue
        bmax = min(bmax, (qlist[j] - (alist[j] * i)) // blist[j])
    allmax = max(allmax, i + bmax)
    i += 1
```
