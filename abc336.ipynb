C - Even Digits

```python
n = int(input()) - 1
if n == 0:
    print(0)
    exit()
acc = ""
i = 0
while n > 5 ** i:
    i += 1
for j in range(i-1,-1,-1):
    d = n//(5**j)
    acc += str(2*d)
    n -= d * (5**j)
print(acc)
```
