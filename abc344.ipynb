A - Spoiler

```python
s = input()
acc = ""
flag = True
for c in s:
    if c == "|":
        flag = not flag
        continue
    if flag:
        acc += c
print(acc)
```

B - Delimiter

```python
import sys
alist = sys.stdin.readlines()
for i in range(len(alist)-1,-1,-1):
  print(alist[i].strip())
```

C - A+B+C

```python
n = int(input())
alist = list(set(map(int, input().split())))
an = len(alist)
m = int(input())
blist = list(set(map(int, input().split())))
bn = len(blist)
l = int(input())
clist = list(set(map(int, input().split())))
cn = len(clist)
q = int(input())
xlist = list(map(int, input().split()))

sset = set()
for i in range(an):
    for j in range(bn):
        for k in range(cn):
            sset.add(alist[i] + blist[j] + clist[k])

for x in xlist:
    if x in sset:
        print("Yes")
    else:
        print("No")
```
