A - Subsegment Reverse

```python
n, l, r = map(int, input().split())
ans = list(range(1, n+1))
ans = ans[:l-1] + list(reversed(ans[l-1:r])) + ans[r:]
print(*ans)
```

B - Nutrients

```python
n,m = map(int, input().split())
alist = list(map(int, input().split()))
acc = [0] * m
for _ in range(n):
    xlist = list(map(int, input().split()))
    for i in range(m):
        acc[i] += xlist[i]
for i in range(m):
    if alist[i] > acc[i]:
        print("No")
        exit()
print("Yes")
```

C - Keys

```python
n,m,k = map(int, input().split())
test_ls = []
rdict = {"o": True, "x": False}
for _ in range(m):
    c,*a,r = input().split()
    test_ls.append([[int(x) - 1 for x in a], rdict[r]])
acc = 0
for i in range(1 << n):
    flag = True
    for t in test_ls:
        cnt = 0
        for j in t[0]:
            if (i >> j) & 1:
                cnt += 1
        if t[1] ^ (cnt >= k):
            flag = False
            break
    if flag:
        acc += 1
print(acc)
```
