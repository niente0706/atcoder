A - Security

```python
s = input()
last = ""
for c in s:
    if c == last:
        print("Bad")
        exit()
    last = c
print("Good")
```

B - Bite Eating

```python
n,l = map(int, input().split())
pie = sum(range(l,l+n))
if l >= 0:
    print(pie - l)
elif l + n - 1 <= 0:
    print(pie - (l + n - 1))
else:
    print(pie)
```

C - Anti-Division

```python
a,b,c,d = map(int, input().split())
euc = [d, c % d]
while euc[1] != 0:
    euc = [euc[1], euc[0] % euc[1]]
else:
    gcd = euc[0]
lcm = c * d // gcd
n = b - a + 1
nc = b // c - (a - 1) // c
nd = b // d - (a - 1) // d
ncd = b // lcm - (a - 1) // lcm
print(n-nc-nd+ncd)
```

D - Megalomania

```python
n = int(input())
i = 1
jobs = []
for i in range(n):
    jobs.append(list(map(int, input().split())))
sorted_jobs = sorted(jobs, key=lambda x: x[1])
t = 0
for j in sorted_jobs:
    if t + j[0] > j[1]:
        print("No")
        exit()
    else:
        t += j[0]
print("Yes")
```
