A - Divisible

```python
n,k = map(int, input().split())
alist = list(map(int, input().split()))
acc = []
for a in alist:
    if a % k == 0:
        acc.append(a // k)
print(*acc)
```

B - Substring

```python
s = input()
acc = []
for i in range(len(s)):
    for j in range(i+1,len(s)+1):
        if s[i:j] not in acc:
            acc.append(s[i:j])
print(len(acc))
```

C - Ideal Holidays

```python
n,a,b = map(int, input().split())
dlist = list(map(int, input().split()))
week = a + b
acc = [week, 0]
for d in dlist:
    if d % week < acc[0]:
        acc[0] = d % week
    if d % week > acc[1]:
        acc[1] = d % week
if acc[1] - acc[0] <= a:
    print("Yes")
else:
    print("No")
```

WA
多分週を跨ぐ場合に間違ってる判定になってる

```python
n,a,b = map(int, input().split())
dlist = list(map(int, input().split()))
week = a + b
sec_max = 0
sec_min = week
for d in dlist[1:]:
    dth = (d - dlist[0]) % week
    if dth >= a:
        dth -= week
    sec_max = max(sec_max, dth)
    sec_min = min(sec_min, dth)
if sec_max - sec_min < a:
    print("Yes")
else:
    print("No")
```
