A - Adjacent Product

```python
n = int(input())
alist = list(map(int, input().split()))
blist = []
for i in range(n-1):
    blist.append(alist[i] * alist[i+1])
print(*blist)
```

C - ╬г

```python
n,k = map(int, input().split())
alist = list(map(int, input().split()))
acc = [True]*k
s = k*(k+1)//2
for a in alist:
    if 1 <= a and a <= k:
        if acc[a-1]:
            s -= a
            acc[a-1] = False
print(s)
```

```python
n,k = map(int, input().split())
alist = list(map(int, input().split()))
alist.sort()
s = k*(k+1)//2
last = 0
for a in alist:
    if a > k:
        print(s)
        exit()
    if a != last:
        last = a
        s -= a
print(s)
```
