```python
t = int(input())
cases = []
i = 0
while i < t:
    cases.append(list(map(int, input().split())))
    i += 1
```

```python
print(t)
print(cases)
```

```python
for c in cases:
    c0 = min(c[0],c[1])
    c1 = max(c[0],c[1])
    if c[2]-c1 > 1 or c[2]-c1 < 0:
        print(0)
        continue
    n0 = 10**(c0)-10**(c0-1)
    n1 = 10**(c1)-10**(c1-1)
    if c[2]-c1 == 1:
        ncase = (n0*(n0+1))//2 + n0*(10**(c0-1)-1)
    elif c0 == c1:
        ncase = (n0-(10**(c0-1)))*(n0-(10**(c0-1))+1)//2
    else:
        ncase = n0*(n1-10**(c0-1)) + (n0*(n0+1))//2
    print(int(ncase) % 998244353)
    
```
