B - Which is ahead?

```python
n = int(input())
pdict = dict(zip(list(map(int, input().split())), range(1,n+1)))
q = int(input())
for _ in range(q):
    a,b = map(int, input().split())
    if pdict[a] < pdict[b]:
        print(a)
    else:
        print(b)
```

C - Many Replacement

```python
n = int(input())
s = input()
q = int(input())
pairs = []
for _ in range(q):
    a,b = input().split()
    pairs.append((a,b))
conv = []
for c in range(ord("a"), ord("z")+1):
    conv.append(chr(c))
for a,b in pairs:
    for i in range(26):
        if conv[i] == a:
            conv[i] = b
acc = ""
base = ord("a")
for c in s:
    acc += conv[ord(c)-base]
print(acc)
```
