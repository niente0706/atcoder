B - Farthest Point

```python
import math
n = int(input())
posls = [list(map(int, input().split())) for _ in range(n)]
def dist(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
for i in range(n):
    temp = [0,i]
    for j in range(n):
        if i == j:
            continue
        if dist(posls[i], posls[j]) > temp[0]:
            temp = [dist(posls[i], posls[j]), j]
    print(temp[1]+1)
```

C - Colorful Beans

TLE

```python
n = int(input())
beans = {}
for _ in range(n):
    a,c = map(int, input().split())
    if c not in list(beans.keys()):
        beans[c] = a
    else:
        if beans[c] > a:
            beans[c] = a
print(max(list(beans.values())))
```

AC

```python
n = int(input())
beans = {}
for _ in range(n):
    a,c = map(int, input().split())
    if c not in beans:
        beans[c] = a
    else:
        beans[c] = min(beans[c], a)
print(max(beans.values()))
```
