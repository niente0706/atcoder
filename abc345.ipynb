```python
s = input()
n = len(s)
for i in range(n):
    if i == 0:
        if s[i] != "<":
            print("No")
            exit()
    elif i == n - 1:
        if s[i] != ">":
            print("No")
            exit()
    elif s[i] != "=":
        print("No")
        exit()
print("Yes")
```

```python
x = int(input())
y = x // 10
if x % 10 == 0:
    print(y)
else:
    print(y + 1)
```

C - One Time Swap

```python
s = input()
n = len(s)
acc = n * (n - 1) // 2
flag = False
for i in range(ord("a"), ord("z") + 1):
    c = s.count(chr(i))
    if not flag and c >= 2:
        flag = True
    acc -= c * (c - 1) // 2
if flag:
    acc += 1
print(acc)
```
