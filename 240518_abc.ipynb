A - Exponential Plant

```python
h = int(input())
plant = 1
i = 1
while plant <= h:
    plant += 2 ** i
    i += 1
print(i)
```

```python
n = int(input())
i = 1
s = []
sum = 0
while i <= n:
    name,score = input().split()
    s.append(name)
    sum += int(score)
    i += 1
print(sorted(s)[sum%n])
```
