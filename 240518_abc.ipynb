A - Exponential Plant

```python
h = int(input())
plant = 1
i = 1
while plant <= h:
    plant += 2 ** i
    i += 1
print(i)
```

```python
n = int(input())
i = 1
s = []
sum = 0
while i <= n:
    name,score = input().split()
    s.append(name)
    sum += int(score)
    i += 1
print(sorted(s)[sum%n])
```

C - AtCoder Magics

```python
n = int(input())
i = 1
cards = {}
while i <= n:
    a,c = map(int, input().split())
    cards[(a,c)] = i
    i += 1
```

```python
acc = []
power = 0
for a,c in sorted(list(cards.keys()), key=lambda x: x[1]):
    if a > power:
        power = a
        acc.append(cards[(a,c)])
ans = ""
for i in sorted(acc):
    ans += str(i) + " "
print(ans[:-1])
```

```python
n = int(input())
i = 1
cards = []
while i <= n:
    a,c = map(int, input().split())
    cards.append([a,c,i])
    i += 1
acc = []
power = 0
for a,c,i in sorted(cards, key=lambda x: x[1]):
    if a > power:
        power = a
        acc.append(i)
print(len(acc))
print(*sorted(acc))
```

D - AtCoder Wallpaper

```python
a,b,c,d = map(int, input().split())
```

```python
x = [(c-a)//4, (c-a)%4, a%4]
y = [(d-b)//2, (d-b)%2, b%2]

area_prof = {(0,0): 2, (0,1): 1, (1,0): 1, (1,1): 2, (2,0): 0, (2,1): 1, (3,0): 1, (3,1): 0}
nx = [x[0], x[0], x[0], x[0]]
for i in range(x[2],x[1]+x[2]):
    nx[i%4] += 1
ny = [y[0], y[0]]
if y[1] == 1:
    ny[y[2]] += 1

area = 0
for i in range(4):
    for j in range(2):
        area += area_prof[(i,j)] * nx[i] * ny[j]
print(area)
```
