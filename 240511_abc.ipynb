```python
n,k = map(int, input().split())
alist = list(map(int, input().split()))
i = 0
open_seat = k
acc = 0
while i < n:
    if alist[i] <= open_seat:
        open_seat -= alist[i]
    else:
        acc += 1
        open_seat = k - alist[i]
    i += 1
acc += 1
print(acc)
```

```python
n = int(input())
alist = sorted(list(map(int, input().split())))
```

```python
print(n)
print(alist)
```

```python
acc = 0
for i in range(n-1):
    for j in range(i+1,n):
        acc += (alist[i] + alist[j]) % 100000000
print(acc)
```

TLE (実行時間超過)だったのでアルゴリズムを改善

```python
acc = sum(alist) * (n - 1)
j = 1
over = 0
for i in range(n):
    while alist[i] + alist[n-j] > 100000000:
        j += 1
    over += j
    if i + j > n:
        break
acc -= over * 100000000
print(acc)
```
