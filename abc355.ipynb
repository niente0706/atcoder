A - Who Ate the Cake?

```python
a,b = map(int, input().split())
sus = [1, 2, 3]
if a == b:
    print(-1)
else:
    sus.pop(max(a,b)-1)
    sus.pop(min(a,b)-1)
    print(sus[0])
```

B - Piano 2

```python
n,m = map(int,input().split())
if n - m > 1:
    print("Yes")
    exit()
alist = list(map(int,input().split()))
blist = list(map(int,input().split()))
asort = sorted(alist)
bsort = sorted(blist)
i = 0
j = 0
flag = False
while i < n and j < m:
    if flag and asort[i] < bsort[j]:
        print("Yes")
        exit()
    elif asort[i] < bsort[j]:
        flag = True
        i += 1
    else:
        flag = False
        j += 1
print("No")
```

C - Bingo 2

```python
n,t = map(int, input().split())
alist = list(map(int, input().split()))
bingo = {}
for i in range(n * n):
    bingo[i] = ((i//n), i%n)
iacc = [0] * n
jacc = [0] * n
naname1 = 0
naname2 = 0
for k in range(t):
    i,j = bingo[alist[k]-1]
    iacc[i] += 1
    if iacc[i] == n:
        print(k+1)
        exit()
    jacc[j] += 1
    if jacc[j] == n:
        print(k+1)
        exit()
    if i == j:
        naname1 += 1
        if naname1 == n:
            print(k+1)
            exit()
    if i + j == n - 1:
        naname2 += 1
        if naname2 == n:
            print(k+1)
            exit()
    
print(-1)
```

D - Intersecting Intervals

```python
n = int(input())
sections = []
for _ in range(n):
    l,r = map(int,input().split())
    sections.append([l-0.1, True])
    sections.append([r, False])
sec_sort = sorted(sections, key = lambda x: x[0])
acc = 0
result = 0
for s in sec_sort:
    if s[1]:
        result += acc
        acc += 1
    else:
        acc -= 1
print(result)
```
