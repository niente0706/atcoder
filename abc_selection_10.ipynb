2: ABC 081 A - Placing Marbles

```python
ss = input()
acc = 0
for s in ss:
    if s == "1":
        acc += 1
print(acc)
```

3: ABC 081 B - Shift Only

```python
n = int(input())
alist = list(map(int, input().split()))
acc = 0
while True:
    for i in range(n):
        if alist[i] % 2 == 1:
            print(acc)
            exit()
        else:
            alist[i] = alist[i] // 2
    acc += 1
```

4: ABC 087 B - Coins

```python
print(100-(100*0)//50)
```

```python
a = int(input())
b = int(input())
c = int(input())
x = int(input())
n = 0
amax = min(a, x//500)
for i in range(amax+1):
    y = x - (500*i)
    bmax = min(b, y//100)
    for j in range(bmax+1):
        if (y - (100*j)) // 50 <= c:
            n += bmax + 1 - j
            break
print(n)
```

5: ABC 083 B - Some Sums

```python
n,a,b = map(int, input().split())
acc = 0
if n == 10000:
    n = 9999
    if a == 1:
        acc += 10000
for i in range(1,n+1):
    s = i // 1000 + (i % 1000) // 100 + (i % 100) // 10 + i % 10
    if a <= s and s <= b:
        acc += i
print(acc)
```

```python
# 模範解答
n,a,b = map(int, input().split())
acc = 0
for i in range(n+1):
    s = 0
    x = i
    while True:
        s += x % 10
        x = x // 10
        if x == 0:
            break
    if a <= s and s <= b:
        acc += i
print(acc)
```
