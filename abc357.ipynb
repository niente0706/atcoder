A - Sanitize Hands

```python
n,m = map(int, input().split())
hlist = list(map(int, input().split()))

i = 0
while m >= 0:
    if i == n:
        print(i)
        exit()
    m -= hlist[i]
    i += 1
print(i-1)
```

B - Uppercase and Lowercase

```python
s = input()
lower = ord("a")
l = 0
u = 0
for c in s:
    if ord(c) >= lower:
        l += 1
    else:
        u += 1
if l < u:
    print(s.upper())
else:
    print(s.lower())
```

C - Sierpinski carpet

```python
n = int(input())
if n == 0:
    print("#")
    exit()
def inc(k):
    n = len(k)
    acc = []
    for i in range(3*n):
        if n <= i < 2*n:
            acc.append(k[i%n] + ("."*n) + k[i%n])
        else:
            acc.append(k[i%n] * 3)
    return acc
k = ["###","#.#","###"]
for i in range(n-1):
    k = inc(k)
for l in k:
    print(l)
```

E - Reachability in Functional Graph

```python
n = int(input())
adict = dict(zip(range(1,n+1), map(int,input().split())))
accall = [0] * n
flag = [True] * n
for i in range(1,n+1):
    if accall[i-1] != 0:
        continue
    acc = [i]
    flag[i-1] = False
    k = i
    while flag[adict[k]-1]:
        if k == adict[k]:
            break
        flag[adict[k]-1] = False
        acc.append(adict[k])
        k = adict[k]
    if accall[adict[k]-1] != 0:
        for l in range(len(acc)):
            accall[acc[l]-1] = len(acc)-l+accall[adict[k]-1]
    else:
        j = acc.index(adict[k])
        for l in range(len(acc)):
            if l < j:
                accall[acc[l]-1] = len(acc)-l
            else:
                accall[acc[l]-1] = len(acc)-j
print(sum(accall))
```
