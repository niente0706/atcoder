A - Scoreboard

```python
n = int(input())
tacc = 0
aacc = 0
for _ in range(n):
    t,a = map(int, input().split())
    tacc += t
    aacc += a
if tacc > aacc:
    print("Takahashi")
elif tacc < aacc:
    print("Aoki")
else:
    print("Draw")
```

B - Extended ABC

```python
s = input()
acc = ["A", "B", "C"]
for c in s:
    if c not in acc:
        print("No")
        exit()
    if c == "B":
        acc = ["B", "C"]
    elif c == "C":
        acc = ["C"]
print("Yes")
```

C - Lining Up 2

```python
n = int(input())
adict = dict(zip(list(range(1,n+1)), list(map(int, input().split()))))
iset = set(adict.keys())
aset = set(adict.values())
last = (iset - aset).pop()
acc = []
while last != -1:
    acc.append(last)
    last = adict[last]
acc.reverse()
print(*acc)
```
