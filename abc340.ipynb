A - Arithmetic Progression

```python
a,b,d = map(int, input().split())
acc = [a]
n = a
while n < b:
    n += d
    acc.append(n)
print(*acc)
```

B - Append

```python
Q = int(input())
q = []
for _ in range(Q):
    q1,q2 = map(int, input().split())
    if q1 == 1:
        q.append(q2)
    else:
        print(q[-q2])
```

C - Divide and Divide

```python
n = int(input())
i = 1
while n // (2 ** i) > 1:
    i += 1
print(n * i + 2 * (n % (2 ** i)))
```
