B - Adjacency Matrix

```python
n = int(input())
for _ in range(n):
    ai = list(map(int, input().split()))
    acc = []
    for i in range(n):
        if ai[i] == 1:
            acc.append(i+1)
    print(*acc)
```

C - 343

```python
n = int(input())

def palindrome(s):
    n = len(s)
    for i in range(n//2):
        if s[i] != s[n-i-1]:
            return False
    return True

def cubed(n):
    acc = []
    i = 1
    while True:
        if i ** 3 > n:
            break
        acc.append(i ** 3)
        i += 1
    return acc

c_ls = cubed(n)
c_ls.reverse()
for c in c_ls:
    if palindrome(str(c)):
        print(c)
        exit()
```
