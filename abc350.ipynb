A - Past ABCs

```python
s = input()
if s[:3] == "ABC":
    if int(s[3:]) <= 349 and int(s[3:]) != 316 and int(s[3:]) != 000:
        print("Yes")
    else:
        print("No")
```

B - Dentist Aoki

```python
n, q = map(int, input().split())
teeth = [0] * n
tlist = list(map(int, input().split()))
for t in tlist:
    teeth[t-1] += 1
acc = 0
for i in range(n):
    if teeth[i-1] % 2 == 0:
        acc += 1
print(acc)
```

C - Sort

```python
test = [3, 4, 1, 2, 5]
i = 0
a = test[i] - 1
if i != a:
    test[i], test[a] = test[a], test[i]
print(test)
```

```python
n = int(input())
alist = list(map(int, input().split()))
acc = []
for i in range(n):
    while i != alist[i] - 1:
        a = alist[i] - 1
        acc.append((i, a))
        alist[i], alist[a] = alist[a], alist[i]
print(len(acc))
for i, j in acc:
    print(i+1, j+1)
```
